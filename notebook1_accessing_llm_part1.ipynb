
Your Python code runs on your computer and uses the SDK to build a request. The SDK sends your prompt and API key securely to the LLM provider’s server, where the key is checked and the model generates an answer. The server then sends that response back to your Python program.


LLM server: the remote server run by the model provider that receives your request, checks it, runs the language model, and sends back the answer.

SDK: a software package or library that helps your code talk to the provider’s API more easily.

API key: a secret credential used to identify and authorize your application when it sends requests to the server.

### First Google 

```python
import os
from google import genai

# Create the Gemini client — pass your API key once, reuse the client forever
client = genai.Client(api_key=GEMINI_API_KEY)


# Send a prompt — gemini-2.0-flash is fast, cheap, and highly capable
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Write me about Mt. Everest in exactly two sentences."
)

# response.text holds the plain text answer — no JSON parsing needed

print("Gemini :", response.text)
```

## Anthropic

```python
import anthropic

PROMPT = "Write me about Mt. Everest in exactly two sentences."

# ── Claude 4 (Anthropic) ─────────────────────────────────────
claude = anthropic.Anthropic(api_key=ANTHROPIC_API_KEY)

msg = claude.messages.create(
    model="claude-opus-4-5",
    max_tokens=200,
    messages=[{"role": "user", "content": PROMPT}]
)
print("Claude :", msg.content[0].text)
```

### OpenAI

```python
from openai import OpenAI

PROMPT = "Write me about Mt. Everest in exactly two sentences."

gpt = OpenAI(api_key=OPENAI_API_KEY)
resp = gpt.chat.completions.create(
    model="gpt-4o",
    max_tokens=200,
    messages=[{"role": "user", "content": PROMPT}]
)
print("GPT-4o :", resp.choices[0].message.content)
```
