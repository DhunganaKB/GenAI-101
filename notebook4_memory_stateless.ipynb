## The Goldfish Problem — Stateless by Default

An LLM is stateless — each API call is completely independent, like meeting a brilliant stranger for the very first time, every single time you call. Stateless means no state is stored between calls — the model has zero memory of what you said one second ago in a different request.

### Key Points

- each API call is completely independent
- no state is stored between calls
- intentional design, not a flaw
- model cannot connect turn one to turn two

> 💡 **Analogy:** Imagine a world-class doctor who loses all memory the moment you leave the room. You return the next day and they have no idea who you are, what you described, or what they prescribed. Brilliant at answering a single question — helpless across a conversation. That is a stateless LLM: incredible power, zero continuity between calls.


## Stateless — Watching the Model Forget

Two separate API calls — the model forgets everything between them

### Step-by-Step Breakdown

**Turn 1 — introduce yourself** — First API call: one message, one question. The model reads 'My name is Sarah' and replies warmly. Everything looks fine. But notice what we do with that reply — we print it and discard it. The model's response is not saved anywhere. The conversation is already gone.

**Turn 2 — a completely new call** — Second API call: a brand new messages list with just one entry — 'What is my name?' This call has no connection to the first one whatsoever. It is like walking up to a different person and asking them your own name. The word 'Sarah' has never appeared in this call, so the model has never seen it.

**The output — complete amnesia** — Turn 1 knows Sarah perfectly. Turn 2 has no idea. Same model, ten seconds apart, total memory loss. This is what stateless looks like in practice. The model is not misbehaving — it is working exactly as designed. Two calls, two completely independent contexts, zero shared memory.



```python
from openai import OpenAI
client = OpenAI("Your API KEY GOES HERE")

# Turn 1 — tell the model your name
r1 = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "My name is Sarah."}
    ]
)
print("Turn 1:", r1.choices[0].message.content)

# Turn 2 — brand new call, no history passed at all
r2 = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "What is my name?"}
    ]
)
print("Turn 2:", r2.choices[0].message.content)
```

## Stateful — Building Memory in 10 Lines

One Python list is all the memory an LLM needs — build it, grow it, pass it every time

### Step-by-Step Breakdown

**history — one list, total memory** — One empty Python list. That is the entire memory system. No database, no file, no special library. Every user message and every model reply will live in this list. Python is the memory. The LLM is just reading whatever you hand to it.

**chat() — append, call, save** — The chat function does three things in order: it appends your message to history, passes the FULL history list to the API — not just the latest message — then saves the model's reply back into history as an assistant message. Next time you call chat(), the history is two messages longer. The model now sees everything that was ever said.

**Three turns — zero forgetting** — Three calls. Turn 3 asks what the very first message was — and the model answers correctly. Not because it has built-in memory. Because that first message is still sitting in the history list and we pass the whole list every time. Your code is the memory. The model is just reading it.



```python
from openai import OpenAI
client = OpenAI("Your API KEY GOES HERE")

history = []   # this list IS the memory

def chat(user_message):
    history.append({"role": "user", "content": user_message})
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=history        # full history every time
    )
    reply = response.choices[0].message.content
    history.append({"role": "assistant", "content": reply})
    return reply

print(chat("My name is Sarah."))
print(chat("What is my name?"))
print(chat("What was the first thing I told you?"))
```

### Key Takeaways

- LLMs are stateless by default — every API call is completely isolated
- stateful means your code builds a history list and passes every prior turn
- memory lives in your code, not in the model
- the ten-line chat() pattern is the foundation of every chatbot
- the context window is the memory limit for very long conversations
